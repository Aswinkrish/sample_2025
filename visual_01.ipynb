```python
# load package

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
# hist plot 
```

```python
# mainly continious variable 
```

```python
data['Age'].hist()
```

```python
data['Age'].hist()
plt.show()
```

```python
data[['Age', 'Sal']].hist()
plt.show()
```

```python
# seaborn 

plt.figure(figsize= (12, 5))

sns.histplot(data= data, x = 'Age')

plt.show()
```

```python
# seaborn 

plt.figure(figsize= (12, 5))

sns.histplot(data= data, y = 'Age', color = 'orange')

plt.show()
```

```python
data.columns
```

```python
plt.figure(figsize= (12, 5))

sns.histplot(data= data, x = 'Age', color = 'orange', kde = True)

plt.show()
```

```python
data['Age-10'] = data['Age'] - 10
```

```python
plt.figure(figsize= (12, 5))

sns.histplot(data= data, x = 'Age', color = 'orange', kde = True)
sns.histplot(data= data, x = 'Age-10', color = 'red', kde = True)

plt.show()
```

```python
plt.figure(figsize= (12, 5))

sns.histplot(data= data, x = 'Age', color = 'orange', kde = True, label = "Age")
sns.histplot(data= data, x = 'Age-10', color = 'red', kde = True, label = "Age-10")

plt.xlabel("Age Vs Age10")
plt.ylabel("Frequency")
plt.title("Age Comparision")
plt.legend()
plt.show()
```

```python
plt.subplots(2,2, figsize = (15, 9))
```

```python
# sub plots - 
```

```python
fig,  axes = plt.subplots(2,2, figsize = (15, 9))

sns.histplot(data= data, x = 'Age', color = 'orange', kde = True, ax = axes[1,1])
sns.histplot(data= data, x = 'Age-10', color = 'red', kde = True, ax = axes[0,0])
sns.histplot(data= data, x = 'JoiningYear', color = 'green', kde = True, ax = axes[0,1])
sns.histplot(data= data, x = 'Education', color = 'yellow', kde = True, ax = axes[1,0])

plt.show()
```

```python
# Scatter Plot 
```

```python
plt.figure(figsize= (15, 6))


sns.scatterplot(data= data, y = "Age", x = data.index)
```

```python
plt.figure(figsize= (15, 6))

data_s = data.head(20)
sns.scatterplot(data= data_s, y = "Age", x = data_s.index)
plt.show()
```

```python
sns.set_style("dark") #{darkgrid, whitegrid, dark, white, ticks}

plt.figure(figsize= (15, 6))

data_s = data.head(20)
sns.scatterplot(data= data_s, y = "Age", x = data_s.index)
plt.show()
```

```python
# adding face color

sns.set_style("dark") #{darkgrid, whitegrid, dark, white, ticks}

fig, ax = plt.subplots(figsize= (15, 6))
ax.set_facecolor("black")

data_s = data.head(20)
sns.scatterplot(data= data_s, y = "Age", x = data_s.index, color = "red")
plt.show()
```

```python
# Line plot
```

```python
# adding face color

sns.set_style("dark") #{darkgrid, whitegrid, dark, white, ticks}

fig, ax = plt.subplots(figsize= (15, 6))
ax.set_facecolor("black")

data_s = data.head(20)
sns.lineplot(data= data_s, y = "Age", x = data_s.index, color = "red", marker = 'o', label = "Age")
sns.lineplot(data= data_s, y = "Age-10", x = data_s.index, color = "green", marker = 'o', label = "Age-10")

plt.legend()
plt.show()
```
